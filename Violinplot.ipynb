```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.violinplot(x = 'day', y= 'total_bill', data= df, palette= 'Accent', hue= 'time')
plt.show()
```

```python
sns.violinplot(x = 'time', y= 'total_bill', data= df, palette= 'Dark2', hue= 'sex', split= True)
plt.show()
```

```python
sns.violinplot(x = 'total_bill', y= 'time', data= df, palette= 'Dark2', hue= 'sex', split= True, orient = 'h')
plt.show()
```

```python
sns.violinplot(x = 'total_bill', y= 'time', data= df, palette= 'Dark2', hue= 'sex', split= True, orient = 'h', inner = 'point')
plt.show()
```
