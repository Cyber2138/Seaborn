```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.set(style = 'whitegrid')
sns.boxplot(x = 'day', y = 'total_bill', data = df, hue = 'time', palette= 'Dark2', showmeans = True, linewidth = 2, meanprops = {'marker':'+', 'markeredgecolor' : 'blue'})
plt.show()
```
